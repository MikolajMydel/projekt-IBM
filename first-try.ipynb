# Połączenie tabeli dc z marvel

```python
marvel_characters_info = pd.read_csv("./characters/marvel_characters_info.csv", index_col = 0)
dc_characters_info = pd.read_csv("./characters/marvel_dc_characters.csv", index_col = 0, encoding = "iso-8859-1")
characters_stats = pd.read_csv("./characters/characters_stats.csv")
```

```python
characters_info = pd.concat([marvel_characters_info, dc_characters_info])

print(len(characters_info["Name"].unique()))
print(len(characters_info["Name"]))
```

## Pozbycie się duplikatów 

```python
characters_info = characters_info.drop_duplicates(subset=['Name'])
```

## Skasowanie niepotrzebnych danych 

```python
characters_info.drop(["Race", "Publisher", "SkinColor", "Height", "Weight", "Identity", "Status", "Appearances", "FirstAppearance", "Year", "Universe"], 
                     axis = 1, inplace=True)
```

```python
characters_info
```

# Połączenie dc i marvela z statystykami

```python
characters_info_status = pd.concat([characters_stats, characters_info])
```

```python
characters_info_status = characters_info_status.drop_duplicates(subset=['Name'])
```

```python
characters_info_status
```

```python
characters_info_status.isna().sum()
```

```python
characters_info_status = characters_info_status.dropna(subset=["Alignment", "Intelligence", "Strength", "Durability", "Power", "Combat"])
characters_info_status.drop(["Gender", "EyeColor", "HairColor"], axis = 1, inplace=True)
```

```python
characters_info_status.to_csv("./characters/characters_info_status.csv")
```
