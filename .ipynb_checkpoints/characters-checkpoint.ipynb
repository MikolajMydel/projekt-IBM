# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data import

```python
characters_stats = pd.read_csv("./characters/characters_stats.csv")
characters_stats.set_index("Name")

marvel_characters_info = pd.read_csv("./characters/marvel_characters_info.csv")
marvel_characters_info.set_index("Name")
```

# Merging tables

```python
characters_info = marvel_characters_info.merge(characters_stats, on="Name", how='left')
```

```python
characters_info.drop("ID", axis = 1, inplace=True)
characters_info.drop(["Alignment_y"], axis = 1, inplace=True)
characters_info.rename(columns={"Alignment_x": "Alignment"}, inplace=True)

characters_info.set_index("Name")
```

# Normalizing alignment column

```python
characters_info["Alignment"] = characters_info["Alignment"].replace(['good', 'bad', '-', 'neutral'], [0,1,0,0])
```

```python
characters_info = characters_info.drop_duplicates(subset=['Name'])
```

# Data visualisation

```python
good, bad = characters_info['Alignment'].value_counts()
plt.style.use('_mpl-gallery')
width = 50

fig,ax=plt.subplots(figsize=(12,7),edgecolor='k')
p1=ax.bar(good, good, width, color='g')
p2=ax.bar(bad, bad, width, color='r')
ax.set_xticks([good, bad])
ax.set_xticklabels(("Good", "Bad"))
ax.set_title('Graph')
plt.xlabel('Alignment')
plt.ylabel('Counts')

for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+20, p.get_height()+10))

plt.grid()
plt.show()
```

```python
characters_info.to_csv("./characters/final_characters.csv")
```
