# 2.5. Strategie uzupełniania brakujących danych

W którymś momencie pracy z danymi napotykamy na problem brakujących danych. Jest to sytuacja dość częsta gdy łączymy ze sobą dane pochodzące z różnych źródeł, jednak zdarza się również w przypadku pojedynczego ich dostawcy. Wystarczy, że system z którego je pobieramy dopuszcza opcjonalne atrybuty.

Istnieje kilka szkół, które mówią o tym w jaki sposób postępować z brakującymi informacjami.

## Brak informacji jest również informacją

Zbiór danych, który otrzymujemy, pochodzi z pewnego systemu i jest gromadzony w ustalony sposób. Fakt, iż dla niektórych obserwacji, część z cech nie jest wypełniona, może wiele mówić o samym procesie zbierania danych bądź też sygnalizować problemy, jeśli spojrzymy na problem globalnie. Każda z obserwacji opowiada jednak zupełnie inną historię - jest to osobny przypadek ze zbioru, np. osobny klient czy zdarzenie, a więc powinniśmy na samym początku przejrzeć jakie mogą być przyczyny braków.

```python
import pandas as pd
```

```python
symptoms_df = pd.read_csv("./data/2_5_symptoms.csv")
symptoms_df
```

## Usuwanie niekompletnych obserwacji

Dość oczywistą strategią radzenia sobie z brakującymi wartościami jest usunięcie elementów, które mają jakiekolwiek atrybuty niewypełnione. W przypadku gdy takich obserwacji jest obiektywnie niewiele, prawdopodobnie nie stracimy za dużo informacji.

```python
symptoms_df.dropna()
```

Jeśli problem dotyczy tylko pojedynczych kolumn, to możemy również usunąć ją całkowicie albo zanotować tylko fakt obecności braku bądź istnienia wartości w danej obserwacji.

```python
symptoms_df.insert(column="have_headache_info", loc=3, 
                   value=symptoms_df["headache"].isna())
symptoms_df
```

## Uzupełnianie brakujących danych

Bardzo często natrafimy jednak na moment, gdy będziemy zmuszeni dokonać uzupełnienia brakujących wartości. W zależności od typu danych będziemy mogli dobrać inne strategie.

### Dane kategoryczne

#### One-hot encoding

W przypadku danych kategorycznych, wykorzystanie metody one-hot encoding, z automatu koduje informację o braku wartości dla danego atrybutu.

```python
pd.get_dummies(symptoms_df)
```

#### Wypełnianie najczęściej występującą wartością

Dość prostym podejściem jest wykorzystanie najczęściej występującej wartości jako wartości oczekiwanej dla danego atrybutu. Takie podejście zakłada niejawnie, że jedna z wartości będzie dominować w zbiorze.

```python
from sklearn.impute import SimpleImputer
```

```python
mf_imputer = SimpleImputer(strategy="most_frequent")
mf_imputer.fit(symptoms_df)

mf_imputer.transform(symptoms_df)
```

```python
pd.DataFrame(mf_imputer.transform(symptoms_df), 
             columns=symptoms_df.columns)
```

#### Wypełnianie stałą wartością

Poszerzenie zbioru dopuszczalnych wartości dla kategorii może być sensowną alternatywą aby zakodować brak informacji w danym polu.

```python
const_imputer = SimpleImputer(strategy="constant", 
                              fill_value="Unknown")
const_imputer.fit(symptoms_df)
```

```python
pd.DataFrame(const_imputer.transform(symptoms_df), 
             columns=symptoms_df.columns)
```

### Dane numeryczne

Dane, które będziemy dostarczać algorytmom do uczenia, będą w dużej mierze numeryczne. Dla tego typu danych istnieje kilka strategii uzupełniania, jednakże dobór odpowiedniej metody powinien zostać poprzedzony dogłębną analizą rzeczywistego rozkładu danych.

```python
weather_df = pd.read_csv("./data/2_5_weather.csv", index_col=0)
weather_df.sample(n=5)
```

```python
weather_df.plot(figsize=(10.5, 7.5), rot=90)
```

```python
weather_df.plot.density(figsize=(10.5, 7.5))
```

#### Uzupełnianie danych średnią/medianą

Dość częstymi strategiami uzupełniania danych numerycznych są strategie oparte o średnią artytmetyczną lub medianę. Scikit-learn umożliwia implementację obydwu możłiwości w podobny sposób jak w poprzednich dwóch przypadkach.

```python
mean_imputer = SimpleImputer(strategy="mean")
mean_imputer.fit(weather_df)
```

```python
pd.DataFrame(mean_imputer.transform(weather_df)) \
    .plot(figsize=(10.5, 7.5), rot=90)
```

```python
median_imputer = SimpleImputer(strategy="median")
median_imputer.fit(weather_df)
```

```python
pd.DataFrame(median_imputer.transform(weather_df)) \
    .plot(figsize=(10.5, 7.5), rot=90)
```

```python
weather_df.describe()
```

#### Niestandardowe rozkłady danych

Rzeczywiste dane mogą nas często zaskoczyć. Poniższy przykład pochodzi z mojego doświadczenia i był dość zaskakujący, kiedy zobaczyłem go na samym początku.

```python
orders_df = pd.read_csv("./data/2_5_orders.csv", index_col=0)
orders_df.sample(n=5)
```

```python
orders_df.plot.density(figsize=(15.0, 7.5),
                       subplots=True, layout=(2, 2), 
                       sharex=False)
```

```python
orders_df.describe()
```

### Zaawansowane metody uzupełniania danych

Machine Learning zakłada przewidywanie danych, których nie posiadamy. W związku z tym, czasem spotkamy się z koniecznośnią uzupełnienia brakujących informacji używając pewnych modeli pomocnicznych, zanim przejdziemy do odpowiedzi na postawiony problem.
